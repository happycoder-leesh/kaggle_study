### 6. 결측치 삭제,대체(dropna, fillna)
- df.dropna() - 결측치 갖는 행을 삭제
- df.fillna() - 결측치를 인자값으로 대체

```python
train = train.dropna()
test = test.fillna(0)
```

```python
train.info()
```

```python
test.info()
```

### 7. scikit-learn 모델링

```python
import sklearn
from sklearn.tree import DecisionTreeRegressor
```

```python
X_train = train.drop(['count'], axis=1)
Y_train = train['count']

model = DecisionTreeRegressor()
model.fit(X_train, Y_train)
```

```python
pred = model.predict(test)
pred[:5]
```

### 8. 제출파일생성
- to_csv() : predict csv 파일로 만들기

```python
submission = pd.read_csv('submission.csv')
submission['count'] = pred
submission.to_csv('sub.csv',index=False)
```
