```python
#  

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
df_copy = df.copy()
df_copy.drop('Region', inplace=True, axis=1)
df_copy.head(10)
```

```python
vif = []
df_val = df_copy.values
for i in range(len(df_copy.columns)):
    vif.append(variance_inflation_factor(df_val, i))
    
vif
```

```python
vif_dataframe = pd.DataFrame()
vif_dataframe['columns'] = df_copy.columns
vif_dataframe['VIF'] = vif
vif_dataframe
```
