```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features = ['enginesize', 'boreratio', 'citympg', 'highwaympg', 'drivewheel_fwd', 'drivewheel_rwd', 'cylindernumber_four']
df_infor[features] = scaler.fit_transform(df_infor[features])
```

```python
df_infor[features].head()
```

```python
from sklearn.model_selection import train_test_split

features = ['enginesize', 'boreratio', 'citympg', 'highwaympg', 'drivewheel_fwd', 'drivewheel_rwd', 'cylindernumber_four']

X = df_infor[features]
y = df_new['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score, KFold

model = XGBRegressor(max_depth=4)
model.fit(X_train, y_train, eval_metric='rmse')
kfold = KFold(n_splits=10, shuffle=True)
kf_cv_scores = cross_val_score(model, X_train, y_train, cv=kfold )
print("K-fold CV average score: %.2f" % kf_cv_scores.mean())
```

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
import numpy as np

pred = model.predict(X_test)
print(np.sqrt(mean_squared_error(pred, y_test)))
```

```python
x_ax = range(len(y_test))
plt.scatter(x_ax, y_test, s=5, color="blue", label="original")
plt.plot(x_ax, y_test, lw=0.8, color="red", label="predicted")
plt.legend()
plt.show()
```
